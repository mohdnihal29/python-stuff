```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_nodes_from([1,2,3])
G.nodes()
```

```python
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,1)
G.edges()
```

```python
G.node[1]['label'] = 'blue'
G.node[2]['label'] = 'red'
G.node[3]['label'] = 'purple'
G.nodes(data=True)
```

```python
nx.draw(G)
plt.show()
```

```python
F = nx.Graph()
F.add_edge('A', 'B', weight=4)
F.add_edge('B', 'D', weight=2)
F.add_edge('A', 'C', weight=3)
F.add_edge('C', 'D', weight=4)

nx.draw(F)
plt.show()
```

```python
nx.shortest_path(F, 'A', 'D', weight="weight")
```

```python
nx.to_dict_of_dicts(F)
```
